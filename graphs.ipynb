```python
genre_columns = ['Action', 'Adventure', 'Animation', 'Comedy', 'Crime', 
                 'Drama', 'Family', 'Fantasy', 'History', 
                 'Horror', 'Music', 'Mystery', 'Romance', 'Science Fiction', 
                 'Thriller', 'War', 'Western']

# Step 1: Sum each genre column to get total movies per genre
genre_counts = df[genre_columns].sum()

# Step 2: Create the bar chart
plt.figure(figsize=(12,6))  # width x height
plt.bar(genre_counts.index, genre_counts.values, color='skyblue')
plt.xticks(rotation=45)  # rotate x-axis labels for readability
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Genre')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Select numeric columns
num_cols = ['vote_average', 'popularity']

# Step 2: Compute correlation matrix
corr = df[num_cols].corr()

# Step 3: Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Vote Average vs Popularity')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    df['popularity_bin'], 
    df['vote_average'], 
    c=df['vote_average'], 
    cmap='coolwarm', 
    alpha=0.7
)
plt.colorbar(label='Vote Average')
plt.xlabel('Popularity Bin')
plt.ylabel('Vote Average')
plt.title('Vote Average vs Popularity Bin (Colored by Rating)')
plt.show()
```
